```python
top_cities = [
    "Mont Saint Michel",
    "St Malo",
    "Bayeux",
    "Le Havre",
    "Rouen",
    "Paris",
    "Amiens",
    "Lille",
    "Strasbourg",
    "Chateau du Haut Koenigsbourg",
    "Colmar",
    "Eguisheim",
    "Besancon",
    "Dijon",
    "Annecy",
    "Grenoble",
    "Lyon",
    "Gorges du Verdon",
    "Bormes les Mimosas",
    "Cassis",
    "Marseille",
    "Aix en Provence",
    "Avignon",
    "Uzes",
    "Nimes",
    "Aigues Mortes",
    "Saintes Maries de la mer",
    "Collioure",
    "Carcassonne",
    "Ariege",
    "Toulouse",
    "Montauban",
    "Biarritz",
    "Bayonne",
    "La Rochelle"
]
```

```python
locations = []
weather = []

for city_id, city in enumerate(top_cities):
  res_gps = requests.get('https://nominatim.openstreetmap.org/search?q={}&format={}'.format(city, 'json'))
  geo_data = res_gps.json()[0]
  latitude = float(geo_data['lat'])
  longitude = float(geo_data['lon'])
  city_id += 1
  location = {
    'location_id': city_id,
    'location': city,
    'location_latitude': latitude,
    'location_longitude': longitude
  }
  locations.append(location)

  res_weather = requests.get('https://api.openweathermap.org/data/3.0/onecall?lat={lat}&lon={lon}&units=metric&exclude={exclude}&appid={API_key}'.format(lat=latitude, lon=longitude, exclude='current,minutely,hourly,alerts', API_key='c2e5d6e560ce0aeec5855494cd7f9266'))
  daily = res_weather.json()['daily']
  for idx,day in enumerate(daily[1:]):
    weather_data = {
      'day': idx + 1,
      'timestamp': day['dt'],
      'location_id': city_id,
    }

    weather_data['temperature'] = day['temp']['day']
    weather_data['weather'] = day['weather'][0]['main']
    weather_data['wind_speed'] = day['wind_speed']
    weather_data['wind_deg'] = day['wind_deg']
    weather_data['pressure'] = day['pressure']
    weather_data['humidity'] = day['humidity']
    weather_data['dew_point'] = day['dew_point']
    weather_data['prob_prcp'] = day['pop']
    weather_data['clouds'] = day['clouds']
    weather_data['uvi'] = day['uvi']
    weather.append(weather_data)
```

```python
pprint(weather[:2])
```

#### Saving Locations

```python
import csv

with open('data/locations.csv', 'w') as location_file:
    fieldnames = ['location_id', 'location', 'location_latitude', 'location_longitude']
    writer = csv.DictWriter(location_file, fieldnames=fieldnames)
    writer.writeheader()
    writer.writerows(locations)
```

#### Saving Weather Data

```python
import json

# Serializing json
weather_json = json.dumps(weather, indent=4)

# Writing to weather.json
with open("data/weather.json", "w") as file:
    file.write(weather_json)
```
